```python
import pandas as pd
import numpy as np
from datetime import datetime as dt
from datetime import timedelta
```

```python
ptt_file = pd.read_csv('PTT.csv', encoding='cp950')
movie_file =pd.read_csv('MOVIES_WITHNick_102_200.csv', encoding='cp950')
```

```python
ptt_file['date'] = ptt_file['date'].apply(pd.to_datetime)
movie_file['DATE_TW'] = movie_file['DATE_TW'].apply(pd.to_datetime)
```

```python
output_rows = []
```

```python
for index, row in movie_file.iterrows():
    start_date = row['DATE_TW']
    end_date = start_date - timedelta(30*6)
    names = row['NICK_NAME'].split('/')
    print(names)
    #select date range
    temp_df = ptt_file[ptt_file.date < start_date]
    temp_df = temp_df[temp_df.date > end_date]
    #select mentioned topic
    if len(names) == 1:
        temp_df = temp_df[temp_df.article_title.str.contains(names[0])]
    elif len(names) == 2:
        temp_df = temp_df[temp_df.article_title.str.contains(names[0]) | 
                         temp_df.article_title.str.contains(names[1])]
    else:
         temp_df = temp_df[temp_df.article_title.str.contains(names[0]) | 
                         temp_df.article_title.str.contains(names[1]) |
                          temp_df.article_title.str.contains(names[2])]
    print(len(temp_df))
    out_row = []
    out_row.append(len(temp_df))
    out_row.append(temp_df['push'].sum())
    out_row.append(temp_df['neutral'].sum())
    out_row.append(temp_df['boo'].sum())
    out_row.append(temp_df['all'].sum())
    output_rows.append(out_row)
```

```python
print(output_rows)
```

```python
import csv

with open("ptt102-200.csv", "w", newline='') as f:
    writer = csv.writer(f)
    writer.writerows(output_rows)
    
```
